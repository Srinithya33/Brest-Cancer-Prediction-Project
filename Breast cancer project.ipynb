```python
df['diagnosis']=df['diagnosis'].map({'M':1,'B':0})
```

```python
df['diagnosis'].unique()
```

```python
x=df.drop('diagnosis',axis=1)
df.head()
```

```python
x.head()
```

```python
y=df['diagnosis']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
df.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
x_train.head(1)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
x_train
```

## MACHINE LEARNING MODELS

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
lr_accu=(accuracy_score(y_test,y_pred))
print(lr_accu)
```

```python
results=pd.DataFrame()
results
```

```python
tempResults=pd.DataFrame({'Algorithm':['Logistic Regression Method'],'Accuracy':[lr_accu]})
results=pd.concat([results,tempResults])
results=results[['Algorithm','Accuracy']]
results
```

# DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
y_pred=dtc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))

```

```python
dtc_accu=accuracy_score(y_test,y_pred)
print(dtc_accu)
```

```python
tempResults=pd.DataFrame({'Algorithm':['Decision tree classifier method'],'Accuracy':[dtc_accu]})
results=pd.concat([results,tempResults])
results=results[['Algorithm','Accuracy']]
results
```

# RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
rfc_accu=accuracy_score(y_test,y_pred)
print(rfc_accu)
```

```python
tempResults=pd.DataFrame({'Algorithm':['Random Forest Classifier Method'],'Accuracy':[rfc_accu]})
results=pd.concat([results,tempResults])
results=results[['Algorithm','Accuracy']]
results
```

# SUPPORT VECTOR CLASSIFIER

```python
from sklearn import svm
svc=svm.SVC()
svc.fit(x_train,y_train)

```

```python
y_pred=svc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
svm_accu=accuracy_score(y_test,y_pred)
print(svm_accu)
```

```python
tempResults=pd.DataFrame({'Algorithm':['Support vector classifier'],'Accuracy':[svm_accu]})
results=pd.concat([results,tempResults])
results=results[['Algorithm','Accuracy']]
results
```
